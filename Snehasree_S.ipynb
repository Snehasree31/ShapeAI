```python
import numpy as num
import pandas as pan
import sklearn
```

```python
from sklearn.datasets import load_boston
a=load_boston()
```

```python
boston=pan.DataFrame(a.data, columns=a.feature_names)
boston.head()
```

```python
boston['MEDV']= a.target
boston.head()
```

```python
boston.isnull()
```

```python
boston.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
m=boston.drop('MEDV',axis=1)
n=boston['MEDV']

m_train, m_test, n_train, n_test =train_test_split (m,n,test_size =0.15, random_state=5)

print(m_train.shape)
print(m_test.shape)
print(n_train.shape)
print(n_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
linear_mod =LinearRegression()
linear_mod.fit(m_train,n_train)
```

```python
n_train_pre =linear_mod.predict(m_train)
RMSE =(num.sqrt(mean_squared_error(n_train, n_train_pre)))
print("The model performance for training set")
print("RMSE is {}\n".format(RMSE))

n_test_pre =linear_mod.predict(m_test)
RMSE=(num.sqrt(mean_squared_error(n_test, n_test_pre)))
print("The model performance for testing set")
print("RMSE is {}".format(RMSE))
```
